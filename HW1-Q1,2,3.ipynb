Q2: Based on visual inspection of the pairs plot, write a function that classifies based on rules

```python
def predict(x1,x2):
    if (x1<0.5 and x2<0.5):
        res_class=2
    elif x1>1 and x2>2:
        res_class=1
    else:
        res_class=3
    return res_class
```

Q3: Apply the rule based classifier in (2) to predict the the test set and report the misclassification rate

```python
 def Misclassification(row):
        x1 = test_dataset['x1'][row]
        x2 = test_dataset['x2'][row]
        if (predict(x1,x2) == test_dataset['class'][row]):
            return 0
        else:
            return 1
```

```python
def MisclassificationRate():
    rate = 0
    num_of_rows = len(test_dataset)
    for row in range(num_of_rows):
        rate+=Misclassification(row)

    return(rate/num_of_rows)
```
